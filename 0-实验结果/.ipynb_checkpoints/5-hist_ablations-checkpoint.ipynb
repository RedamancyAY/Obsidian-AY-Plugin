```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.colors import ListedColormap
from matplotlib.lines import Line2D
from sklearn.manifold import TSNE

plt.rc("font", family="Times New Roman")
plt.rcParams['hatch.color'] = 'white'
plt.rcParams['font.size'] = 13
ROOT_PATH = "/home/ay/data/DATA/1-model_save/00-Deepfake/1-df-audio"
```

```python
from utils import read_test_result
```

```python
models = [
    "AASIST",
    "LCNN",
    "RawNet2",
    "Wave2Vec2",
    # "Ours/ResNet",
]


DATA = []
tasks = ["LibriSeVoc_cross_method", "DECRO_english"]
version = 0
for model in models:
    for task in tasks:
        _data = read_test_result(model, task, version, metric_prefix="test")
        res = _data.dropna().reset_index(drop=True)

        if task == "LibriSeVoc_cross_method":
            res.loc['Avg', :] = res.mean()
            res = res.reset_index(drop=True)
            res = res.iloc[-1:]
        elif task == "DECRO_english":
            res = res.iloc[1:2]
        res["model"] = model
        res["task"] = task
        DATA.append(res)
data = pd.concat(DATA).reset_index(drop=True)
```

```python
Lib_data = data[data['task'] == "LibriSeVoc_cross_method"]
DECRO_data = data[data['task'] == "DECRO_english"]
```

```python
EER_metrics = {
    "eer_Lib": [1 - x for x in Lib_data["test-eer"]],
    "eer_DECRO": [1 - x for x in DECRO_data["test-eer"]],
}
AUC_metrics = {"auc_en": Lib_data["test-auc"], "auc_ch": DECRO_data["test-auc"]}
```
