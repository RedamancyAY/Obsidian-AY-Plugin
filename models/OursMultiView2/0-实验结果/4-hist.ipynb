```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.colors import ListedColormap
from matplotlib.lines import Line2D
from sklearn.manifold import TSNE

plt.rc("font", family="Times New Roman")
plt.rcParams['hatch.color'] = 'white'
plt.rcParams['font.size'] = 13
ROOT_PATH = "/home/ay/data/DATA/1-model_save/00-Deepfake/1-df-audio"
```

```python
from utils import read_test_result
```

```python
models = [
    "AASIST",
    "LCNN",
    "RawNet2",
    "Wave2Vec2",
    "WaveLM",
    "LibriSeVoc",
    "Wav2Clip",
    "AudioClip",
    "RawGAT",
    "Ours/ResNet",
]

DATA = []
task = "LibriSeVoc_cross_dataset"
version = 0
for model in models:
    _data = read_test_result(model, task, version, metric_prefix="test")
    if _data is None:
        continue
    res = _data.dropna().reset_index(drop=True)
    res = res.iloc[8:10]
    res["model"] = model
    res["method"] = ["en", "ch"]
    DATA.append(res)
data = pd.concat(DATA)

en_data = data.query('method == "en"')
ch_data = data.query('method == "ch"')
```

```python
EER_metrics = {
    "eer_en": [1 - x for x in en_data["test-eer"]],
    "eer_ch": [1 - x for x in ch_data["test-eer"]],
}
AUC_metrics = {"auc_en": en_data["test-auc"], "auc_ch": ch_data["test-auc"]}
```

```python
width = 0.32  # the width of the bars
multiplier = 0
fig, ax = plt.subplots(layout="constrained",figsize=(5,5))
x = np.arange(10)
colors = ["red", "blue"]
colors = ["#9489fa", "#f06464","#779043","#624c7c"] 
hatches = ["//", "oo"]
for metric, v in EER_metrics.items():
    offset = width * multiplier + 0.2
    rects = ax.barh(
        x + offset,
        v,
        width,
        label=metric,
        color=colors[multiplier],
        hatch=hatches[multiplier],
        edgecolor='#252626'
    )
    shown_labels = ["{:.2f}".format((1 - a) * 100) for a in v]
    ax.bar_label(rects, shown_labels, padding=1, label_type='edge', fontsize=11)
    multiplier += 1

ax.margins(y=0.01, tight=True)
ax.set_yticks(x+width, [a.split("/")[0] for a in models])
# ax.set_ylim(0.1,10)
ax.set_xticks(np.arange(11) / 10)
# ax.legend(loc='lower right')
ax.legend(['EN', 'CH'], loc=(0.7, 0.15))
ax.set_xlim(0.4, 1)
ax.set_xticklabels(["%.2f"%((1 - a / 10) * 100) for a in np.arange(11)])
# ax.set_xticklabels([f"{int(100 - a*10)}" for a in np.arange(11)])
ax.set_xlabel("EER (%) scores")

# Remove x, y Ticks
ax.xaxis.set_ticks_position("none")
ax.yaxis.set_ticks_position("none")

# Add padding between axes and labels
# ax.xaxis.set_tick_params(pad = 5)
# ax.yaxis.set_tick_params(pad = 10)
for s in ["top", "bottom", "left", "right"]:
    ax.spines[s].set_visible(False)

# Add x, y gridlines
ax.grid(color="grey", linestyle="-", linewidth=1.0, alpha=0.2)
# for i in ax.patches:
#     plt.text(i.get_width()+0.01, i.get_y()+0.23,
#              str(round((1- i.get_width())*100, 2)),
#              fontsize = 10, fontweight ='bold',
#              color ='grey')
# Show top values
ax.invert_yaxis()
plt.savefig('pngs/hist_en_ch.pdf', dpi=500, bbox_inches="tight")

plt.show()
```
