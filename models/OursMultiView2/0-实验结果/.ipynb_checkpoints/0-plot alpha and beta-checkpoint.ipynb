```python
from utils import read_test_result
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()
# Fixing random state for reproducibility
np.random.seed(19680801)
import pandas as pd
```

```python
data = pd.DataFrame()
v = 0
for alpha in [0.5, 0.6, 0.7, 0.8, 0.9, 0.95, 1.0]:
    for beta in [0.5, 0.6, 0.7, 0.8, 0.9, 0.95, 1.0]:


        _data = read_test_result(model="OursMultiView2/ablations/ab", task="ASV2021_inner", version=v)
        v += 1
        if _data is None:
            continue
        
        data.loc[alpha, beta] = _data.iloc[1:].mean()['test-eer'] * 100
```

```python
data
```

```python
title = 'Average EER scores'
fmt = 'png'
sns.heatmap(data.T, annot=True, fmt='.4g')
plt.title(title)
plt.xlabel(r"$\alpha$")
plt.ylabel(r"$\beta$")
plt.savefig(f'imgs/ab.pdf', format='pdf', dpi=300);
```
