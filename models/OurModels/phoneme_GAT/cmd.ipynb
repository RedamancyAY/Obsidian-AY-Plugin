```python
tasks = [
    "ASV2019_LA",
    # "ASV2021_LA",
    'ASV2021_inner',
    "MLAAD_cross_lang",
]

test = 0
gpu = 0
versions = range(0, 3)
for i, ablation in enumerate([
    "no_clip",
    "no_aug",
    "no_GAT",
    # "no_aug_no_GAT",
    "no_pool",
    'edges0',
    'edges2',
    'edges4',
    'edges6',
    'edges8',
    'edges12',
    'edges14',
    'edges16',
    'edges18',
    # "no_pool_no_clip",
    # "no_pool_no_clip_use_raw",
    # "no_pool_no_clip_use_raw_wav2vec",
    "wav2vec",
]):
    gpu = i // 4
    for task in tasks:
        if not test:
            optional = ' --min_epoch 5 ' if task == "MLAAD_cross_lang" else ""
            for v in versions:
                print(f"python train.py --gpu {gpu} --cfg 'OursPhonemeGAT/{task}' --ablation 'ablations/{ablation}' {optional} -v {v};\\")
                print(f"python train.py --gpu {gpu} --cfg 'OursPhonemeGAT/{task}' --ablation 'ablations/{ablation}' -t 1 -v {v};\\")
                if task == "ASV2021_inner":
                    print(f"python train.py --gpu {gpu} --cfg '{model}/{task}'  -t 1 -v {v} --test_noise 1 --test_noise_level 20 --test_noise_type 'bg';\\")

        else:
            for v in versions:
                print(f"python train.py --gpu 0 --cfg 'OursPhonemeGAT/{task}' --ablation 'ablations/{ablation}' -t 1 -v {v};\\")
    print("#", "---" * 30, ";\\")
```
